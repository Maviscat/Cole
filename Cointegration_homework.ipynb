```python
from statsmodels.tsa.stattools import coint
list = ['TCB', 'MBB']
dict = {}
cointe_dict = {}
for element in list:
  A_stock_name = element
  stock_A = stock_historical_data(symbol=A_stock_name, start_date="2021-01-01", end_date='2022-01-01')['Close'].to_numpy()
  dict[element] = stock_A
for element in list:
  if element not in cointe_dict:
    cointe_dict[element] = ['cointegrate with']
  A_stock_name = element
  stock_A = stock_historical_data(symbol=A_stock_name, start_date="2021-01-01", end_date='2022-01-01')['Close'].to_numpy()
  for element2 in list:
    if element2 != element:
      B_stock_name = element2
      stock_B = stock_historical_data(symbol=B_stock_name, start_date="2021-01-01", end_date='2022-01-01')['Close'].to_numpy()
      score, p_value, _ = coint(stock_A, stock_B)
      if p_value < 0.05:
        cointe_dict[element].append(element2)
print(dict) #lưu trữ giá
print(cointe_dict) #danh sách cointegrate
```
